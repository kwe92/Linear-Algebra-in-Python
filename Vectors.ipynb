# Vectors

- Vectors can be made with lists or numpy arrays. numpy arrays are faster to manipulate
- They are homogeneous and have operations meant for manipulation
- Numpy arrays can be computationally manipulated easily

```python
# 2-d vector
v2 = np.array([3.0,-2.0],dtype=float)

# 2-d vector
v2_list = [5,7]

# 3-d vector
v3 = np.array([4, -3, 2], dtype=int)
```

```python
v2 + 4
```

```python
v2_list + 4
```

```python
# Row Vector
v3 * np.array([3,5,6])
```

```python
v3.reshape(-1,1)
```

```python
# turn row vector v3 into a column vector
# Many machine learning alogrithms use column vectors instead of row vectors
# In the case of a vector you can reshape it programatically with array.reshape(-1,1)

v3.reshape(-1,1) * np.array([3,5,6])
```

```python
# Creating a 3d plot
fig = plt.figure(figsize=[10,10])
ax = plt.axes(projection='3d')
ax.plot([0,v3[0]],[0,v3[1]],[0,v3[2]], marker='o', color='blue')
ax.plot([0,0],[0,0],[-4,4],'k')
ax.plot([0,0],[-4,4],[0,0],'k')
ax.plot([-4,4],[0,0],[0,0],'k') 
```

# Vector Additon & Subtraction

```python
v1 = np.array([3,-2],dtype=int)
v2 = np.array([2,4],dtype=int)

print('v1 = {}'.format(v1))
print('v2 = {}'.format(v2))
```

```python
v3 = v1 + v2
```

```python
print(f'v3 = {v3}')
```

```python
plt.plot([0,v1[0]],[0,v1[1]],'b', label='v1', marker='o')
plt.plot([0,v2[0]],[0,v2[1]],'y', label='v2',marker='o')
plt.plot([0,v3[0]],[0,v3[1]],'r', label='v3',marker='o')
# plt.plot([0,v2[0]] + v1[0],[0,v2[1] + v1[1]],'r')
plt.plot([-6,6],[0,0],'black')
plt.plot([0,0],[-6,6],'black')
plt.axis('square')
plt.axis((-6,6,-6,6))
plt.grid()
```

```python
plt.plot([0,v1[0]] + v2[0],[0,v1[1]] + v2[1],'b', label='v1', marker='o')
plt.plot([0,v2[0]],[0,v2[1]],'y', label='v2',marker='o')
plt.plot([0,v3[0]],[0,v3[1]],'r', label='v1 + v2',marker='o')
plt.plot([-6,6],[0,0],'k')
plt.plot([0,0],[-6,6],'k')
plt.legend()
plt.axis('square')
plt.axis((-6,6,-6,6))
plt.grid()
```
