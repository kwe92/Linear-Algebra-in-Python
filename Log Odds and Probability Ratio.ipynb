```python
logOdds = np.array([log(i / (100 - i)) for i in random.choices(np.arange(1,100),k=10000)])
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.histplot(logOdds, bins=12)
ax2 = ax.twinx()
sns.kdeplot(logOdds, color="purple", ax=ax2)
plt.show()
```

```python
class Account:
    
    def __init__(self,name,balance = 0):
        self.name = name
        self.balance = balance
        
    def deposit(self,deposit):
        if deposit > 0:
            self.balance += deposit
            print("New Balance: {}".format(self.balance))
        else:
            print("Can not deposit a negative value!")
        
    def withdraw(self,withdraw):
        if withdraw <= self.balance:
            self.balance -= withdraw
            print("New Balance: {}".format(self.balance))
        else:
            print("Can not withdraw more than current balance: {}".format(self.balance))
    
            
```

```python
acct1 = Account("Kweayon Clark", 10000)
```
